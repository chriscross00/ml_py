- Bayesian classification
    * $$P(L|features) = \frac{P(features|L)P(L)}{P(features)}$$
        * Between 2 labels $L_1 \text{ and } L_2$:
        $$\frac{P(L_1|features)}{P(L_2|features)} = \frac{P(features|L_1)P(L_1)}{P(features|L_2)P(L_2)}$$

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# code settings
pd.set_option('display.max_columns', None)

# visualization settings
plt.rc('figure', figsize=(8,8))
sns.set_style('darkgrid')
```

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(100, 2, centers=2, random_state=2, cluster_std=1.5)


sns.scatterplot(X[:, 0], X[:, 1], hue=y)
```

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(X, y)
```

```python
rng = np.random.RandomState(0)
Xnew = [-6, -14] + [14, 18] * rng.rand(2000, 2)
ynew = model.predict(Xnew)
```

```python
sns.scatterplot(X[:, 0], X[:, 1], hue=y, s=50)
sns.scatterplot(Xnew[:, 0], Xnew[:, 1], hue=ynew, s=25, alpha=0.25)
```

```python
yprod = model.predict_proba(Xnew)
yprod[0:10].round(4)
```
