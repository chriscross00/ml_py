```python
def check_drop_na(df):
    """
    Checks if the dataframe has NANs and if true removes the rows where present
    
    Args:
        df: A dataframe with or without NANs
    Returns:
        df: A dataframe without NANs
    """
    checked = df.isnull().values.any()
    if checked:
        df.dropna(inplace=True)
    return df
```
